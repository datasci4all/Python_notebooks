```python
#### Calculate Similarity Using Distance ####

# Importing necessary libraries
import pandas as pd
from plotnine import ggplot, aes, geom_point, xlab, ylab, ggtitle
from scipy.spatial import distance_matrix

# Import data
small_iris = pd.read_csv("small_iris.csv")

# Plot original data
(ggplot(small_iris, aes(x = "Sepal.Length", y = "Sepal.Width")) +
  geom_point(size = 4) +
  xlab("Sepal Length") +
  ylab("Sepal Width") +
  ggtitle("Sepal Width vs. Sepal Length")
)
```

```python
small_iris_num = small_iris[["Sepal.Length" , "Sepal.Width"]]

# Compute distances
dist_matrix = pd.DataFrame(distance_matrix(small_iris_num, small_iris_num))
dist_matrix
```
