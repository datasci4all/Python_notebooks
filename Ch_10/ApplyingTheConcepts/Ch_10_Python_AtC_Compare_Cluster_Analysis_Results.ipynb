```python
#### Compare Cluster Analysis Results ####

# Importing necessary libraries
import pandas as pd
from plotnine import ggplot, aes, geom_point, xlab, ylab, ggtitle, scale_x_continuous, scale_y_continuous, labs
from sklearn.cluster import KMeans

# Import data
data = pd.read_csv("iris.csv")

# Original Data in all black
(ggplot(data, aes(x = "Sepal.Length", y = "Sepal.Width")) +
  geom_point(size = 5) +
  ylab("Sepal Width") +
  xlab("Sepal Length") +
  ggtitle("Sepal Width vs. Sepal Length")
)
```

```python
## 3 cluster analysis
k_res = KMeans(n_clusters = 3, random_state = 0,
              algorithm = "lloyd").fit(data[["Sepal.Length", "Sepal.Width", "Petal.Length", "Petal.Width"]])

data["clusters3"] = [str(x) for x in k_res.labels_]


# Plot 3 cluster analysis
(ggplot(data, aes(x = "Sepal.Length", y = "Sepal.Width")) +
  geom_point(aes(fill = "clusters3"), color = "black", size = 5) +
  ylab("Sepal Width") +
  xlab("Sepal Length") +
  labs(fill = "Cluster") +
  ggtitle("Sepal Width vs. Sepal Length")
)
```

```python
## 4 cluster analysis
k_res = KMeans(n_clusters = 4, random_state = 0,
              algorithm = "lloyd").fit(data[["Sepal.Length", "Sepal.Width", "Petal.Length", "Petal.Width"]])

data["clusters4"] = [str(x) for x in k_res.labels_]

# Plot 4 cluster analysis
(ggplot(data, aes(x = "Sepal.Length", y = "Sepal.Width")) +
  geom_point(aes(fill = "clusters4"), color = "black", size = 5) +
  ylab("Sepal Width") +
  xlab("Sepal Length") +
  labs(fill = "Cluster") +
  ggtitle("Sepal Width vs. Sepal Length")
)
```
