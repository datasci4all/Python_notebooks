```python
#### Detect Anomalies ####

# Importing necessary libraries
import numpy as np
import pandas as pd

x = np.array([35, 12, 50, 42, 600, 5, 70, 37, 90, 73, 45, 100, 22])

# Compute mean and standard deviation
m = x.mean()
std_dev = x.std(ddof=1)

# Compute z-scores (standardized scores)
z = (x-m)/std_dev

# Organize information into a data frame
standardized_scores = pd.DataFrame({
    'purchases': x,
    'scores': z
})
print(standardized_scores)

```
