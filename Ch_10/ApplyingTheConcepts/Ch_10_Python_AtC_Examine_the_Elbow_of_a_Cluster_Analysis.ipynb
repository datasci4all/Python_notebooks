```python
#### Examine the Elbow of a Cluster Analysis ####

# Importing necessary libraries
import pandas as pd
from plotnine import ggplot, aes, geom_line, xlab, ylab, ggtitle, geom_point
from sklearn.cluster import KMeans

# Import data
data = pd.read_csv("iris.csv")

k_vals = [2, 3, 4, 5, 6, 7, 8, 9, 10]
var_ratio = [0, 0, 0, 0, 0, 0, 0, 0, 0]

# K-means cluster analysis for different k
j = 0
for i in k_vals:
    k_res = KMeans(n_clusters = i, random_state = 0,
              algorithm = "lloyd").fit(data[["Sepal.Length", "Sepal.Width", "Petal.Length", "Petal.Width"]])
    var_ratio[j] = k_res.inertia_
    j = j + 1

# Organize information into a data frame
data_res = pd.DataFrame(k_vals, var_ratio)

# Plot K-means cluster analysis results
(ggplot(data_res, aes(x = "k_vals", y = "var_ratio")) +
  geom_point(size = 5) +
  geom_line() +
  xlab("Number of Clusters (k)") +
  ylab("Within-Cluster Variability") +
  ggtitle("Within-Cluster Variability vs. Number of Clusters")
)
```
