```python
#### Clean Incorrect Formats ####

# Importing necessary libraries
import pandas as pd

# Loading the dataset
data = pd.read_csv('birdwatcher.csv')

# Count the initial number of 'sparrow', 'woodpecker' and 'hummingbird' responses before cleaning
initial_sparrow_count = data['Favorite_Bird'].eq('sparrow').sum()
initial_woodpecker_count = data['Favorite_Bird'].eq('woodpecker').sum()
initial_hummingbird_count = data['Favorite_Bird'].eq('hummingbird').sum()

# Clean the data
data['Favorite_Bird'] = data['Favorite_Bird'].replace(
    {
        r'sparrows': 'sparrow',
        r'woodpeckers': 'woodpecker',
        r'hummingbirds': 'hummingbird',
        r's.*w': 'sparrow',
        r'w.*r': 'woodpecker',
        r'h.*d': 'hummingbird'
    }, regex=True)

# Count the final number of 'sparrow', 'woodpecker' and 'hummingbird' responses after cleaning
final_sparrow_count = data['Favorite_Bird'].eq('sparrow').sum()
final_woodpecker_count = data['Favorite_Bird'].eq('woodpecker').sum()
final_hummingbird_count = data['Favorite_Bird'].eq('hummingbird').sum()

print(f"Initial sparrow count: {initial_sparrow_count}")
print(f"Initial woodpecker count: {initial_woodpecker_count}")
print(f"Initial hummingbird count: {initial_hummingbird_count}")
print(f"Final sparrow count: {final_sparrow_count}")
print(f"Final woodpecker count: {final_woodpecker_count}")
print(f"Final hummingbird count: {final_hummingbird_count}")
```
