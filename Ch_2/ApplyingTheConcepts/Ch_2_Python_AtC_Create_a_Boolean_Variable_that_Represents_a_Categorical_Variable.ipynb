```python
#### Create a Boolean Variable that Represents a Categorical Variable ####

# Import necessary libraries
import pandas as pd

# Load the zoo dataset
zoo = pd.read_csv('Zoo.csv')

# Create boolean variables representing categorical variables
# 'Mammal' is True if class_type is 1, False otherwise
zoo['Mammal'] = False
zoo['Mammal'][zoo['class_type'] == 1] = True

# 'Bird' is True if class_type is 2, False otherwise
zoo['Bird'] = False
zoo['Bird'][zoo['class_type'] == 2] = True

# 'Reptile' is True if class_type is 3, False otherwise
zoo['Reptile'] = False
zoo['Reptile'][zoo['class_type'] == 3] = True

# Count the values of the boolean variables
mammal_count = zoo['Mammal'].value_counts()
bird_count = zoo['Bird'].value_counts()
reptile_count = zoo['Reptile'].value_counts()

# Display the counts
print('Mammal Counts:\n', mammal_count)
print('Bird Counts:\n', bird_count)
print('Reptile Counts:\n', reptile_count)
```
