```python
#### Create a Categorical Variable from a Quantitative Variable ####

# Importing necessary libraries
import pandas as pd

# Read the AmesHousing data file into a DataFrame
AmesHousing = pd.read_csv('AmesHousing.csv')

# Categorize SalePrice
AmesHousing["SalePriceCategory"] = ""
AmesHousing["SalePriceCategory"][AmesHousing["SalePrice"] <= 100000] = "Cheap"
AmesHousing["SalePriceCategory"][(AmesHousing["SalePrice"] > 100000) & (AmesHousing["SalePrice"] <= 150000)] = "Cheap Moderate"
AmesHousing["SalePriceCategory"][(AmesHousing["SalePrice"] > 150000) & (AmesHousing["SalePrice"] <= 200000)] = "Expensive Moderate"
AmesHousing["SalePriceCategory"][AmesHousing["SalePrice"] > 200000] = "Expensive"

# Get the count of each category
print(AmesHousing['SalePriceCategory'].value_counts())
```
