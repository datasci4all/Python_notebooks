```python
#### Identify Missing Data ####

# Importing necessary libraries
import pandas as pd

# Read the data file into a DataFrame
ch2_ac1 = pd.read_csv('Titanic-Dataset.csv')

# Count the number of missing values in GPA
missing_age = ch2_ac1['Age'].isnull().sum()

# Count the total number of rows in the dataset
total_rows = ch2_ac1.shape[0]

# Calculate the percentage of missing GPA values
percentage_missing_age = (missing_age / total_rows) * 100

# Display the results
print(f'Number of missing age values: {missing_age}')
print(f'Total number of rows: {total_rows}')
print(f'Percentage of missing age values: {percentage_missing_age:f}%')
```
