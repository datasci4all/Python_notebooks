```python
#### Impute Missing Data from Internal Data ####

# Importing necessary libraries
import pandas as pd

# Read the data file into a DataFrame
ch2_ac2 = pd.read_csv('World_Happiness_Report_2005-2021.csv')

# Calculate mean and median of 'Healthy life expectancy at birth'
mean_life_expec = ch2_ac2['Healthy life expectancy at birth'].mean()
median_life_expec = ch2_ac2['Healthy life expectancy at birth'].median()

# Impute missing values with mean and median
ch2_ac2['life_expec_mean'] = ch2_ac2['Healthy life expectancy at birth'].fillna(mean_life_expec)
ch2_ac2['life_expec_median'] = ch2_ac2['Healthy life expectancy at birth'].fillna(median_life_expec)

# Calculate the means of the imputed columns and round the results
ch2_ac2[['life_expec_mean', 'life_expec_median']].mean()
```
