```python
#### Splitting a Variable into Multiple Variables ####

# Import necessary libraries
import pandas as pd
from datetime import datetime

# Load the birdwatcher dataset
birdwatcher_data = pd.read_csv('birdwatcher.csv')

# Comment: Isolating the month from the Date variable
# Assuming the date is in a standard format (e.g., YYYY-MM-DD)
birdwatcher_data['Month'] = birdwatcher_data['Date'].apply(lambda x: datetime.strptime(x, '%Y-%m-%d').month)

# OR Separate 'Date' column into three columns for month, day, year
birdwatcher_data[['Year', 'Month', 'Day']] = birdwatcher_data['Date'].str.split('-', n = 2, expand = True)

# Comment: Counting the number of entries for each month
monthly_counts = birdwatcher_data['Month'].value_counts().sort_index()

# Display the monthly counts
print('Entries per Month:\n', monthly_counts)
```
