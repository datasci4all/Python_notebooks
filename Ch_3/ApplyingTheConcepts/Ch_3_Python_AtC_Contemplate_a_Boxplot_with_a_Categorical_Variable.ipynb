```python
#### Contemplate a Box Plot with a Categorical Variable ####

# Importing necessary libraries
import pandas as pd
from plotnine import ggplot, aes, geom_boxplot, xlab, ggtitle, ylab

# Import the data
countries = pd.read_csv("countries.csv")
continents = ["Africa", "Asia", "Europe", "Oceania"]
countries = countries[countries["continent"].isin(continents)]

# Plot the data
(ggplot(countries, aes(x = "continent", y = "whr_score", fill = "continent")) +
  geom_boxplot() +
  xlab("Continent") +
  ylab("World Happiness Report Score") +
  ggtitle("World Happiness Report Score by Continent")
)
```
