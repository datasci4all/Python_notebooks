```python
#### Describe a Bubble Chart ####

# Importing necessary libraries
import pandas as pd
from plotnine import ggplot, aes, geom_point, xlab, ggtitle, ylab, labs

# Import the data
countries = pd.read_csv("countries.csv")

# Plot the data
(ggplot(countries, aes(x = "mean_years_of_schooling", y = "whr_score", size = "population")) +
  geom_point(color = "blue", alpha = .75) +
  xlab("Mean Years of Schooling") +
  ylab("World Happiness Score") +
  labs(size = "Population") +
  ggtitle("World Happiness Score by Mean Years of Schooling")
)
```
