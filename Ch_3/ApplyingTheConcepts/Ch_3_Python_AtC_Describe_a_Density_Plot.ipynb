```python
#### Describe a Density Plot ####

# Importing necessary libraries
import pandas as pd
from plotnine import ggplot, aes, geom_density, xlab, ggtitle

# Import the data
countries = pd.read_csv("countries.csv")

# Plot the data
(ggplot(countries, aes(x = "under_5_mortality_rate")) +
  geom_density() +
  xlab("Under-5 Mortality Rate") +
  ggtitle("Under-5 Mortality Rate Worldwide")
)
```
