```python
#### Explore a Line Chart ####

# Importing necessary libraries
import pandas as pd
from plotnine import ggplot, aes, geom_line, xlab, ggtitle, ylab

# Import the data
life_expectancy = pd.read_csv("lex_long.csv")
life_expectancy = life_expectancy[(life_expectancy["country"] == "USA") &
                                  (life_expectancy["year"] >= 1880) &
                                  (life_expectancy["year"] <= 2020)]

# Plot the data
(ggplot(life_expectancy, aes(x = "year", y = "life_expectancy")) +
  geom_line() +
  xlab("Year") +
  ylab("Life Expectancy") +
  ggtitle("United States Life Expectancy Over Time")
)
```
