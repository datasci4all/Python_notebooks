```python
#### Interpret a Heatmap ####
import pandas as pd
from plotnine import ggplot, aes, geom_tile, xlab, ggtitle, ylab, geom_text, scale_fill_gradient, labs

countries = pd.read_csv("../content/countries.csv")

incomes = ["High income", "Upper middle income",
                             "Lower middle income"]
country_names = ['Argentina', 'Bolivia', 'Brazil',
                        'Colombia', 'Chile','Ecuador','French Guiana','Guyana',
                        'Paraguay', 'Peru', 'Suriname',
                        'Trinidad and Tobago', 'Uruguay', 'Venezuela']
countries = countries[countries["income_group"].isin(incomes)]
countries = countries[countries["country"].isin(country_names)]

income_order = pd.Categorical(countries["income_group"], categories=["High income", "Upper middle income",
                                                                    "Lower middle income"])
countries = countries.assign(income_order = income_order)

(ggplot(countries, aes(x = "income_order", y = "country", fill = "ggei")) +
  geom_tile() +
  geom_text(aes(label = "ggei")) +
  scale_fill_gradient(low = "darksalmon", high = "deepskyblue") +
  xlab("Income Group") +
  ylab("Country") +
  labs(fill = "GGEI") +
  ggtitle("")
)
```
