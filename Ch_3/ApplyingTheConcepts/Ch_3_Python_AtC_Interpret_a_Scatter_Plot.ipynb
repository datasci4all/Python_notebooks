```python
#### Interpret a Scatter Plot ####

# Importing necessary libraries
import pandas as pd
from plotnine import ggplot, aes, geom_point, xlab, ggtitle, ylab, geom_smooth

# Import the data
countries = pd.read_csv("countries.csv")

# Plot the data
(ggplot(countries, aes(x = "gni_per_capita", y = "life_expectancy")) +
  geom_point() +
  geom_smooth(method = "lm", se = False, color = "blue") +
  xlab("GNI per capita") +
  ylab("Life Expectancy") +
  ggtitle("Life Expectancy by GNI per capita")
)
```
