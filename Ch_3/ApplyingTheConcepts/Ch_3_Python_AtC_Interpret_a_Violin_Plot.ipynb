```python
#### Interpret a Violin Plot ####

# Importing necessary libraries
import pandas as pd
from plotnine import ggplot, aes, geom_density, xlab, ylab, ggtitle, geom_violin, geom_boxplot, scale_x_continuous

def no_labels(values):
    return [""] * len(values)

# Import the data
countries = pd.read_csv("countries.csv")

# Plot the data
(ggplot(countries, aes(x = 1, y = "gni_per_capita")) +
  geom_violin(color = "black", fill = "salmon") +
  geom_boxplot(width = 0.1) +
  xlab("") +
  ylab("GNI Per Capita") +
  scale_x_continuous(labels=no_labels) +
  ggtitle("Distribution of GNI Per Capita Worldwide")
)
```
