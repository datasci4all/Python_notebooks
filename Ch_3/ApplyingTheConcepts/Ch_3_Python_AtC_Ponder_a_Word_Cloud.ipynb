```python
#### Ponder a Word Cloud ####
import pandas as pd
from wordcloud import WordCloud
import matplotlib.pyplot as plt

spam_texts = pd.read_csv("../content/spam_text_messages.csv")

d = {}
for index, row in spam_texts.iterrows():
    d[row["word"]] = row["freq"]
word_cloud = WordCloud(background_color = "white").generate_from_frequencies(d)
plt.imshow(word_cloud)
plt.axis("off")
plt.show()
```
