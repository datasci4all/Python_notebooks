```python
#### Work with a Boxplot ####

# Importing necessary libraries
import pandas as pd
from plotnine import ggplot, aes, geom_boxplot, ylab, ggtitle, xlab, scale_x_continuous

def no_labels(values):
    return [""] * len(values)

# Import the data
countries = pd.read_csv("countries.csv")

# Plot the data
(ggplot(countries, aes(x = 1, y = "mean_years_of_schooling")) +
  geom_boxplot(fill = "purple", color = "black") +
  ylab("Mean Years of Schooling") +
  xlab("") +
  scale_x_continuous(labels=no_labels) +
  ggtitle("Distribution of Mean Years of Schooling Worldwide")
)
```
