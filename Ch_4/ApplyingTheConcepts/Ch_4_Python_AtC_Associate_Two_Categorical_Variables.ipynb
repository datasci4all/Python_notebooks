```python
#### Associate Two Categorical Variables ####

# Importing necessary libraries
import pandas as pd
from plotnine import ggplot, aes, geom_bar, scale_fill_manual, labs

# Load the dataset
imdb_df = pd.read_csv('IMDb.csv')

# Filter for the specified genres
selected_genres = ['Drama', 'Action', 'Animation', 'Comedy']
filtered_df = imdb_df[imdb_df['Genre'].isin(selected_genres)]

# Find the top four Maturity Ratings in terms of counts
top_maturity_ratings = filtered_df['Certificate'].value_counts().nlargest(4).index
top_maturity_ratings_df = filtered_df[filtered_df['Certificate'].isin(top_maturity_ratings)]

# Define a color-blind safe palette
color_blind_palette = ['#d55e00', '#cc79a7', '#0072b2', '#f0e442']

# Create a bar chart for the top four Maturity Ratings using a color-blind safe palette
(ggplot(top_maturity_ratings_df, aes(x = "Genre", fill = "Certificate")) +
    geom_bar(position = "dodge") +
    scale_fill_manual(values = color_blind_palette) +
    labs(title = 'Side-by-Side Column Chart of Categorical Variable Certificate\nby Categorical Variable Genre.',
         x = 'Genre', y = 'Count')
)
```
