```python
#### Calculate Mean, Mode, Median, and 90% Trimmed Mean ####

# Importing necessary libraries
import pandas as pd
from scipy.stats import trim_mean

# Reading the data from 'ch4_ac1.csv' into a DataFrame 'ch4_ac1'
summarystats_db = pd.read_csv("IMDb.csv")

# Calculating mean, median, mode, and trimmed mean for 'User_Score'
mean_rating = summarystats_db['Meta_score'].mean()
median_rating = summarystats_db['Meta_score'].median()
mode_rating = summarystats_db['Meta_score'].mode()[0]
mean_90_rating = trim_mean(summarystats_db['Meta_score'].dropna(), proportiontocut=0.05)

# Creating a DataFrame to display the calculated statistics
summary = pd.DataFrame([{
    "Mean Rating": mean_rating,
    "90% Trimmed Mean Rating": mean_90_rating,
    "Median Rating": median_rating,
    "Mode Rating": mode_rating
}])

# Displaying the summary DataFrame
print(summary)
```
