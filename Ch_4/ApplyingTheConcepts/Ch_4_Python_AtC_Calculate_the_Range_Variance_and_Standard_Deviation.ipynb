```python
#### Calculate the Range, Variance, and Standard Deviation ####

# Importing necessary libraries
import pandas as pd

# Load the dataset
videogames_df = pd.read_csv('IMDb.csv')

# Calculating range, variance, and standard deviation for 'Meta_score'
range_meta_score = videogames_df['Meta_score'].max() - videogames_df['Meta_score'].min()
variance_meta_score = videogames_df['Meta_score'].var()
std_dev_meta_score = videogames_df['Meta_score'].std()

# Output the calculated values
print("Range of Meta score:", range_meta_score)
print("Variance of Meta score:", variance_meta_score)
print("Standard Deviation of Meta score:", std_dev_meta_score)
```
