```python
#### Describe the Shape of the Histogram ####

# Importing necessary libraries
import pandas as pd
from plotnine import ggplot, aes, geom_histogram, labs, theme_minimal

# Load the dataset
imdb_df = pd.read_csv('IMDb.csv')

# Plotting a histogram of User_Ratings
(ggplot(imdb_df, aes(x = "IMDB_Rating")) +
    geom_histogram(bins = 10, fill = '#0072b2', color = 'black') +
    labs(title = 'Histogram of IMDb Rating', x = 'Meta Score', y = 'Frequency') +
    theme_minimal()
)
```
