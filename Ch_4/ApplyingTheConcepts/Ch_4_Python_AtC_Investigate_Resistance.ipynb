```python
#### Investigate Resistance ####

# Importing necessary libraries
import pandas as pd
from scipy import stats

# Load the dataset
imdb_df = pd.read_csv('IMDb.csv')

# Calculating various statistics for 'Gross'
mean_Gross = imdb_df['Gross'].mean()
trimmed_mean_Gross = stats.trim_mean(imdb_df['Gross'], 0.05)
median_Gross = imdb_df['Gross'].median()
mode_Gross = imdb_df['Gross'].mode()[0]
range_Gross = imdb_df['Gross'].max() - imdb_df['Gross'].min()
variance_Gross = imdb_df['Gross'].var()
std_dev_Gross = imdb_df['Gross'].std()

# Creating a table to display the results
summary_stats = pd.DataFrame({
    'Statistic': ['Mean', '80% Trimmed Mean', 'Median', 'Mode', 'Range', 'Variance', 'Standard Deviation'],
    'Value': [mean_Gross,
              trimmed_mean_Gross,
              median_Gross,
              mode_Gross,
              range_Gross,
              variance_Gross,
              std_dev_Gross]
})

print("Original Data Summary Statistics:")
print(summary_stats)

# Adding an outlier of 10000 to the 'Gross' column
imdb_df_with_outlier = imdb_df.copy()
imdb_df_with_outlier = imdb_df_with_outlier.append({'Gross': 10000}, ignore_index=True)

# Recalculating statistics with the outlier
mean_Gross_outlier = imdb_df_with_outlier['Gross'].mean()
trimmed_mean_Gross_outlier = stats.trim_mean(imdb_df_with_outlier['Gross'], 0.05)
median_Gross_outlier = imdb_df_with_outlier['Gross'].median()
mode_Gross_outlier = imdb_df_with_outlier['Gross'].mode()[0]
range_Gross_outlier = imdb_df_with_outlier['Gross'].max() - imdb_df_with_outlier['Gross'].min()
variance_Gross_outlier = imdb_df_with_outlier['Gross'].var()
std_dev_Gross_outlier = imdb_df_with_outlier['Gross'].std()

# Creating a table to display the results with the outlier
summary_stats_outlier = pd.DataFrame({
    'Statistic': ['Mean', '80% Trimmed Mean', 'Median', 'Mode', 'Range', 'Variance', 'Standard Deviation'],
    'Value': [mean_Gross_outlier,
              trimmed_mean_Gross_outlier,
              median_Gross_outlier,
              mode_Gross_outlier,
              range_Gross_outlier,
              variance_Gross_outlier,
              std_dev_Gross_outlier]
})

print("\nSummary Statistics with Outlier:")
print(summary_stats_outlier)
```
