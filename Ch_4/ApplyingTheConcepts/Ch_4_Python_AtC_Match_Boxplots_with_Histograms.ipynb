```python
#### Match Boxplots with Histograms ####

# Importing necessary libraries
import pandas as pd
from plotnine import ggplot, aes, geom_boxplot, labs, facet_wrap, geom_histogram

# Load the dataset
videogames_df = pd.read_csv('IMDb.csv')

# Filter the dataset
filtered_genres = ['Drama', 'Animation','Comedy']
videogames_filtered = videogames_df[videogames_df['Genre'].isin(filtered_genres)]

# Boxplots of Runtime
(ggplot(videogames_filtered, aes(x="Genre", y="Runtime")) +
  geom_boxplot() +
  labs(title='Boxplots of Runtime for Drama, Animation, and Comedy',
       x='Genre', y='Runtime')
)
```

```python
# Histograms of Runtime
(ggplot(videogames_filtered, aes(x="Runtime")) +
  geom_histogram(binwidth=1) +
  facet_wrap("Genre") +
  labs(title='Histograms of Runtime for Drama, Animation, and Comedy')
)
```
