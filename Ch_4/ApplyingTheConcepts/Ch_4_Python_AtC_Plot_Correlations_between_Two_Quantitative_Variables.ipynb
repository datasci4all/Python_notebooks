```python
#### Plot Correlations between Two Quantitative Variables ####

# Importing necessary libraries
import pandas as pd
from plotnine import ggplot, aes, geom_point, geom_smooth, annotate, labs, theme_bw

# Load the dataset
imdb_df = pd.read_csv('IMDb.csv')

# Calculate the correlation coefficient
corr_coefficient = imdb_df[['Meta_score', 'Gross']].corr().iloc[0, 1]

# Create a scatterplot with a regression line
(ggplot(imdb_df, aes(x="Meta_score", y="Gross")) +
  geom_point() +
  geom_smooth(method='lm', color='red', se=False) +
  annotate('text', x=imdb_df['Meta_score'].min(), y=imdb_df['Gross'].max(), label=f'Corr: {corr_coefficient:.2f}', size=7) +
  labs(title='Scatterplot of Meta_score and Gross') +
  theme_bw(base_size = 14)
)
```

```python
# Calculate the correlation coefficient
corr_coefficient = imdb_df[['Runtime', 'IMDB_Rating']].corr().iloc[0, 1]

# Create a scatterplot with a regression line
(ggplot(imdb_df, aes(x="Runtime", y="IMDB_Rating")) +
  geom_point() +
  geom_smooth(method='lm', color='red', se=False) +
  annotate('text', x=imdb_df['Runtime'].max(), y=imdb_df['IMDB_Rating'].min(), label=f'Corr: {corr_coefficient:.2f}', size=7) +
  labs(title='Scatterplot of Runtime and IMDB_Rating') +
  theme_bw(base_size = 16)
)
```
