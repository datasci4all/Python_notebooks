```python
#### Create a New Variable ####

# Importing necessary libraries
import pandas as pd

# Load the CSV file into a DataFrame
products_df = pd.read_csv('Products.csv')

# Create a new DataFrame with 'Price' and a calculated column 'Price_in_Turkish_Lira'
turk_conversion_rate = 30
products_df['Price_in_Turkish_Lira'] = products_df['Price'] * turk_conversion_rate

# Set the option to display all rows (change 1000 to your desired number)
pd.set_option('display.max_rows', 1000)

# Display the resulting DataFrame
display(products_df.style.hide(axis="index"))
```
