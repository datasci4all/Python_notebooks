```python
#### Filter Missing Values ####

# Importing necessary libraries
import pandas as pd

# Load the CSV file into a DataFrame
orderdetails_df = pd.read_csv('OrderDetails.csv')

# Filter the DataFrame for rows where 'OrderID' is missing (NaN or None)
filtered_orderdetails_missing_postalcode = orderdetails_df[orderdetails_df['OrderID'].isnull()]

# Calculate the number of observations
num_observations = len(filtered_orderdetails_missing_postalcode)

# Print the number of observations
print(f"Number of observations where Order is null: {num_observations}")
```
