```python
#### Filter Multiple Conditions Simultaneously ####

# Importing necessary libraries
import pandas as pd

# Load the CSV file into a DataFrame
orderdetails_df = pd.read_csv('OrderDetails.csv')

# Filter the DataFrame for rows where 'OrderID' is 10293 and 'Quantity' is 10 or higher
filtered_orderdetails_df = orderdetails_df[(orderdetails_df['OrderID'] == 10293 ) & (orderdetails_df['Quantity']>=10)]

# Calculate the number of observations
num_observations = len(filtered_orderdetails_df)

# Print the number of observations
print(f"Number of observations where Order ID = 10293 and Quantity >= 10: {num_observations}")

# Display the resulting DataFrame with all columns
display(filtered_orderdetails_df.style.hide(axis = "index"))

# Add a few blank lines
print("\n\n")


# Load the CSV file into a DataFrame
orderdetails_df2 = pd.read_csv('OrderDetails.csv')
# Filter the DataFrame for rows where 'OrderID' is 10293 and 'Quantity' is greater than 15
filtered_orderdetails_df2 = orderdetails_df2[(orderdetails_df2['OrderID'] == 10294 ) & (orderdetails_df['Quantity'] > 15)]

# Calculate the number of observations
num_observations = len(filtered_orderdetails_df2)

# Print the number of observations
print(f"Number of observations where Order ID = 10293 and Quantity > 15: {num_observations}")

# Display the resulting DataFrame with all columns
display(filtered_orderdetails_df2.style.hide(axis = "index"))
```
