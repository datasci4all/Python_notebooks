```python
#### Filter a Quantitative Variable ####

# Importing necessary libraries
import pandas as pd

# Load the CSV file into a DataFrame
orderdetails_df = pd.read_csv('OrderDetails.csv')

# Filter the DataFrame for rows where 'Quantity' is greater than or equal to 45
filtered_orderdetails_df45 = orderdetails_df[orderdetails_df['Quantity'] >= 45]

# Calculate the number of observations
num_observations = len(filtered_orderdetails_df45)

# Print the number of observations
print(f"Number of observations where the Pirce >= 45: {num_observations}")

# Display the resulting DataFrame with all columns
display(filtered_orderdetails_df45.style.hide(axis = "index"))

# Add a few blank lines
print("\n\n")



# Filter the DataFrame for rows where 'Quantity' is less than 10
filtered_orderdetails_df10 = orderdetails_df[orderdetails_df['Quantity'] > 10]

# Calculate the number of observations
num_observations = len(filtered_orderdetails_df10)

# Print the number of observations
print(f"Number of observations where the Quantity > 10: {num_observations}")

# Display the resulting DataFrame with all columns
display(filtered_orderdetails_df10.style.hide(axis = "index"))
```
