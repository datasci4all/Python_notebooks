```python
#### Merge Two Table ####

# Importing necessary libraries
import pandas as pd

# Load the CSV files into DataFrames (replace with your file paths)
products_df = pd.read_csv('Products.csv')
categories_df = pd.read_csv('Categories.csv')

# Perform an inner join between 'Products' and 'Categories' on 'CategoryID'
result_df = pd.merge(products_df[['CategoryID', 'ProductName']], categories_df[['CategoryID', 'CategoryName']], on='CategoryID', how='inner')

# Set the option to display all rows
pd.set_option('display.max_rows', 1000)

# Calculate the number of observations
num_observations = len(result_df)

# Print the number of observations
print(f"Number of observations after joining: {num_observations}")

# Display the resulting DataFrame with only 'ProductName' and 'CategoryName' columns
display(result_df[['ProductName', 'CategoryName']].style.hide(axis="index"))
```
