```python
#### Summarize a Quantitative Variable ####

# Importing necessary libraries
import pandas as pd

# Load the CSV file into a DataFrame
orderdetails_df = pd.read_csv('OrderDetails.csv')

# Calculate the average price
summary_stat = orderdetails_df['Quantity'].mean()

# Format the result with two decimal places
formatted_summary_stat = summary_stat

# Print the formatted result
print(f"Maximum Quantity: {formatted_summary_stat}")

# Add a few blank lines
print("\n\n")

# Load the CSV file into a DataFrame
orderdetails_df = pd.read_csv('OrderDetails.csv')

# Calculate the average price
summary_stat2 = orderdetails_df['Quantity'].max()

# Format the result with two decimal places
formatted_summary_stat2 = summary_stat2

# Print the formatted result
print(f"Maximum Quantity: {formatted_summary_stat2}")

# Add a few blank lines
print("\n\n")

# Load the CSV file into a DataFrame
orderdetails_df = pd.read_csv('OrderDetails.csv')

# Calculate the average price
summary_stat3 = orderdetails_df['Quantity'].min()

# Format the result with two decimal places
formatted_summary_stat3 = summary_stat3

# Print the formatted result
print(f"Minimum Quantity: {formatted_summary_stat3}")
```
