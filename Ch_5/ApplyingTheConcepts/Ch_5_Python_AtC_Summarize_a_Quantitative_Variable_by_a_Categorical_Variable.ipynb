```python
#### Summarize a Quantitative Variable by a Categorical Variable ####

# Importing necessary libraries
import pandas as pd

# Load the CSV file into a DataFrame
products_df = pd.read_csv('OrderDetails.csv')

# Group the DataFrame by 'ProductID' and calculate the average price for each group
summary_stat_by_product = products_df.groupby('ProductID')['Quantity'].agg('mean')

# Reset the index to have 'ProductID' as a column in the result
summary_stat_by_product = summary_stat_by_product.reset_index()

# Rename the columns to match the SQL query
summary_stat_by_product.columns = ['ProductID', 'AverageQuantity']

# Display the resulting DataFrame
display(summary_stat_by_product.style.hide(axis = "index"))


# Group the DataFrame by 'ProductID' and calculate the average price for each group
summary_stat_by_order = products_df.groupby('OrderID')['Quantity'].agg('mean')

# Reset the index to have 'ProductID' as a column in the result
summary_stat_by_order = summary_stat_by_order.reset_index()

# Rename the columns to match the SQL query
summary_stat_by_order.columns = ['OrderID', 'AverageQuantity']

# Display the resulting DataFrame
display(summary_stat_by_order.style.hide(axis = "index"))
```
