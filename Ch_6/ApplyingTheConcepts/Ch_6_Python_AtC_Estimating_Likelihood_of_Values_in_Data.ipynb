```python
#### Estimating Likelihood of Values in Data ####

# Importing necessary libraries
import pandas as pd
from plotnine import ggplot, aes, geom_histogram, ggtitle, xlab, ylab

# Import data
penguins = pd.read_csv("penguins.csv")

# Plot data
(ggplot(penguins, aes(x = 'body_mass_g')) +
  geom_histogram() +
  ylab("Count") +
  xlab("Body Mass (g)") +
  ggtitle("Body Mass of Palmer Penguins")
)
```

```python
# Compute proportions
print('Proportion of penguins with body mass between 3500 and 5500')
print(len(penguins['body_mass_g'][(penguins['body_mass_g'] <= 5500)
                                  & (penguins['body_mass_g'] >= 3500)])/len(penguins))

print('Proportion of penguins with body mass greater than or equal to 6000')
print(len(penguins['body_mass_g'][penguins['body_mass_g'] >= 6000])/len(penguins))
```
