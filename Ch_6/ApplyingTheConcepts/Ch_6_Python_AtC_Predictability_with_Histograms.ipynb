```python
#### Predictability from Histogram ####

# Importing necessary libraries
import pandas as pd
from plotnine import ggplot, aes, geom_histogram, facet_wrap, labs

# Import data
skeeball = pd.read_csv("skeeball.csv")

# Plot data
(ggplot(skeeball, aes(x = "scores", fill = "person")) +
  geom_histogram() +
  facet_wrap("person", nrow = 3) +
  labs(
    fill = "Gamer",
    x = "Skeeball Scores"
  )
)
```

```python
# Compute average scores
print("Average skeeball Scores:")
print(skeeball.groupby("person").mean())

# Compute proportions
print("Proportion of Darc's Scores Over 400:")
print(sum(skeeball['scores'][skeeball['person'] == "Darc"] > 400)/1000)
print("Proportion of Qpawnz's Scores Over 400:")
print(sum(skeeball['scores'][skeeball['person'] == "Qpawnz"] > 400)/1000)
print("Proportion of Ember's Scores Over 400:")
print(sum(skeeball['scores'][skeeball['person'] == "Ember"] > 400)/1000)
```
