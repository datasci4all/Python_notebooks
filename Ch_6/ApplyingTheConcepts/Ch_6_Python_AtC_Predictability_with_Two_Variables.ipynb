```python
#### Predictability with Two Variables ####

# Importing necessary libraries
import pandas as pd
from plotnine import ggplot, aes, geom_point, ylim, labs, ggtitle

# Import data
treats = pd.read_csv("treats_hour_age.csv")

# Plot treats vs. day_hour
(ggplot(treats, aes(x = 'day_hour', y = 'treats')) +
  geom_point(size = 2.5) +
  ylim(0, 60) +
  labs(
    x = "Hour of the Day",
    y = "Total Number of Treats"
  ) +
  ggtitle("Total Treats vs. Hour of the Day")
)
```

```python
# Plot treats vs. age
(ggplot(treats, aes(x = 'age', y = 'treats')) +
  geom_point(size = 2) +
  ylim(0, 60) +
  labs(
    x = "Age (Weeks)",
    y = "Total Number of Treats"
  ) +
  ggtitle("Total Treats vs. Age")
)
```
