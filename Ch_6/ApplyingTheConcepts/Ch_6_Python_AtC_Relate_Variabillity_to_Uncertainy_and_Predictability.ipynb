```python
#### Relate Variability to Uncertainty and Predictability ####

# Importing necessary libraries
import pandas as pd
import random
import numpy as np
from plotnine import ggplot, aes, geom_bar, facet_wrap, ylab, xlab, ggtitle, scale_y_continuous
random.seed(1234)

# Perform simulations
n = 10000
x3 = random.choices(np.arange(1, 21), k = n)
x = random.choices(np.arange(1, 21), weights = [*[0.005263158]*19, .9], k = n)
x2 = random.choices(np.arange(1, 21), weights = [*[0.01470588]*17, .25, .25, .25], k = n)

# Organize results
d = pd.DataFrame({
  'rolls' : [*x, *x2, *x3],
  'die' : [*["Die 1"]*n, *["Die 2"]*n, *["Die 3"]*n]
})

# Plot results
(ggplot(d, aes(x = 'rolls')) +
  geom_bar() +
  facet_wrap('die', nrow = 3) +
  scale_y_continuous(breaks = [0, 2500, 5000, 7500],
                     labels = ["0.00", "0.25", "0.50", "0.75"]) +
  ylab("Proportion of 100,000 Rolls") +
  xlab("Roll Outcome") +
  ggtitle("Rolls for Three Different Dice")
)
```
