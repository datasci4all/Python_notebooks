```python
# Importing necessary libraries
import pandas as pd
import scipy.stats as stats

# Load the data
school_data = pd.read_csv('school.csv')

# Perform a one-sample t-test
# Null Hypothesis: The mean number of DailyFruitVeg is 2.21
# Alternative Hypothesis: The mean DailyFruitVeg is less than 2.21
sample_mean = school_data['DailyFruitVeg'].mean()
population_mean = 2.21
sample_std = school_data['DailyFruitVeg'].std()
n = len(school_data)

t_statistic, p_value = stats.ttest_1samp(school_data['DailyFruitVeg'], 2.21)

# Adjusting the p-value for a one-tailed test
p_value /= 2

# Outputting the results
print("T-statistic:")
print(t_statistic)

print("P-value:")
print(p_value)
```
