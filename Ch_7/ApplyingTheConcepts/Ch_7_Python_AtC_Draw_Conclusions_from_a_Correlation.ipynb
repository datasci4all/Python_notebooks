```python
# Importing necessary libraries
import pandas as pd
import random
import numpy as np
from plotnine import ggplot, aes, geom_histogram, facet_wrap, ylab, xlab, ggtitle, labs, geom_vline
from scipy.stats import pearsonr

# Load the dataset
data = pd.read_csv('icecream1.csv')

# Extracting ice_cream_sales and number_drownings
ice_cream_sales = data['ice_cream_sales']
number_drownings = data['number_drownings']

# Calculating the Pearson correlation coefficient and p-value
correlation_coefficient, x = pearsonr(ice_cream_sales, number_drownings)

# Output the results
print("Observed correlation coefficient:")
print(correlation_coefficient)
```

```python
# Perform simulation-based hypothesis test

# Create data frame to store simulations in
sim_n = 1000
d = pd.DataFrame({
  'n' : [0]*sim_n,
})

null_value = 0

# Perform simulations
k = 0
for i in np.arange(sim_n):
    simcor, x = pearsonr(ice_cream_sales, random.sample(sorted(number_drownings), k = len(number_drownings)))

    d['n'][k] = simcor
    k = k + 1

# Plot simulations
(ggplot(d, aes(x = 'n')) +
  geom_histogram(alpha = .55, bins = 20) +
  geom_vline(xintercept = correlation_coefficient, color = "red") +
  labs(
      x = "Sample Correlations",
      y = "Count"
  ) +
  ggtitle("Distribution of Sample Correlations")
)
```

```python
# Compute and print p-value
print("P-value:")
print(sum(d['n'] > correlation_coefficient)/sim_n)
```
