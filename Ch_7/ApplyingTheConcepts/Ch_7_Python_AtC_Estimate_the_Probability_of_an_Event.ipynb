```python
# Importing necessary libraries
import pandas as pd
import numpy as np
from plotnine import ggplot, aes, geom_histogram, labs

# Setting the seed for reproducibility
np.random.seed(777)

# Parameters for the die-rolling simulation
n = 100  # Number of rolls in each simulation
sides = 6  # Number of sides on the die
target = 6  # The specific outcome we're interested in (e.g., rolling a 6)

# Null Distribution Simulation
num_sims = 1000
target_counts = [np.sum(np.random.choice(range(1, sides+1), size=n) == target) for _ in range(num_sims)]
target_data = pd.DataFrame({'count_of_target': target_counts})

# Define the bin edges so that they are centered around the integers
bin_edges = np.arange(min(target_counts) - 0.5, max(target_counts) + 1.5, 1)

# Plotting the null distribution with centered bins
(ggplot(target_data, aes(x = 'count_of_target')) +
 geom_histogram(bins = 23, fill = "blue", color = "black") +
 labs(
     x = 'Number of Times Rolling a 6',
     y = 'Count',
     title = 'Counts of Rolling a Six with a Six-Sided Die\nafter 1000 Trials of 100 Die Rolls'
 ))
```

```python
# Compute number and proportion of times at least 20 sixes were seen
print("Number of Times at least 29 sixes were seen in 1000 trials of 100 dice rolls")
print(sum(target_data['count_of_target'] == 29))


print("Proportion of Times at least 29 sixes were seen in 1000 trials of 100 dice rolls")
print(sum(target_data['count_of_target'] == 29)/1000)
```
