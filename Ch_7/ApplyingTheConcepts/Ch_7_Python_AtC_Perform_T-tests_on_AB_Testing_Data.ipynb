```python
# Importing necessary libraries
import pandas as pd
from scipy.stats import ttest_ind

# Load the dataset
data = pd.read_csv('portfolio.csv')  # Replace with the actual file path

# Separate the data into two groups based on the image seen
A_group = data[data['treatment'] == 'A']
B_group = data[data['treatment'] == 'B']

# Calculate the number of participants in each group
num_A_participants = A_group.shape[0]
num_B_participants = B_group.shape[0]

# Calculate the mean session duration for each group
mean_session_A = A_group['view_time'].mean()
mean_session_B = B_group['view_time'].mean()

# Perform t-tests for session duration and contacted portion
t_stat_session, p_value_session = ttest_ind(A_group['view_time'], B_group['view_time'], equal_var=True)

# Output the results
print("Number of Participants - A Image:", num_A_participants)
print("Number of Participants - B Image:", num_B_participants)
print("Mean Session Duration - A Image:", mean_session_A)
print("Mean Session Duration - B Image:", mean_session_B)
print("P-Value for Session Duration:", p_value_session)
```
