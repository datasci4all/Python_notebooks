```python
# Importing necessary libraries
import pandas as pd
import scipy.stats as stats

# Load the data
vb_data = pd.read_csv('volleyball.csv')

# Perform a one-sample t-test
sample_mean = vb_data['hit_perc'].mean()
population_mean = 30
sample_std = vb_data['hit_perc'].std()
n = len(vb_data)

t_statistic, p_value = stats.ttest_1samp(vb_data['hit_perc'], 30)

# Adjusting the p-value for a one-tailed test
p_value /= 2

# Calculating the mean and standard deviation
mean_hit_perc = vb_data['hit_perc'].mean()
std_hit_perc = vb_data['hit_perc'].std()

# Outputting the results
print("Mean of hit_perc:")
print(mean_hit_perc)

print("Standard Deviation of hit_perc:")
print(std_hit_perc)

print("T-statistic:")
print(t_statistic)

print("P-value:")
print(p_value)
```
