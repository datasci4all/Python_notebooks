```python
#### Reading a Decision Tree ####

# Importing necessary libraries
import pandas as pd
from sklearn import tree
from sklearn.pipeline import Pipeline
from sklearn.tree import DecisionTreeClassifier, plot_tree

# Import data
data = pd.read_csv("college.csv")

# Build decision tree
tree_pipeline = Pipeline(
  [("tree", DecisionTreeClassifier(max_depth=1))]
).set_output(transform="pandas")

X_train = data.drop('Private', axis=1)
y_train = data['Private'].astype('category').cat.codes

# Fit and plot tree
tree_fitted = tree_pipeline.fit(X_train, y_train)
tree.plot_tree(tree_fitted['tree'], feature_names = list(X_train.columns),
                filled = True, class_names = sorted(list(data['Private'].unique()))
    )
```
