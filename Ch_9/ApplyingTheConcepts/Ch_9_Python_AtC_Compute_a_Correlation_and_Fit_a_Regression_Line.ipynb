```python
#### Compute a Correlation and Fit a Regression Line ####

# Importing necessary libraries
import pandas as pd
from plotnine import ggplot, aes, geom_point
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LinearRegression
import numpy as np

# Import data
carseats = pd.read_csv("carseats.csv")
carseats = carseats.dropna(subset = ["Sales", "Advertising"])

# Plot the data
(ggplot(carseats, aes(x = "Advertising", y = "Sales")) +
  geom_point()
)
```

```python
# Compute and print the correlation
print("Correlation:")
np.corrcoef(carseats["Sales"], carseats["Advertising"])
```

```python
# Regression model
lr_pipeline = Pipeline(
  [("lr", LinearRegression(fit_intercept = True))]
).set_output(transform="pandas")

X_train = pd.DataFrame(carseats["Advertising"])
y_train = carseats['Sales']
lr_fitted = lr_pipeline.fit(X_train, y_train)

print("Regression Slope Coefficient:")
print(lr_fitted["lr"].coef_[0])
```
