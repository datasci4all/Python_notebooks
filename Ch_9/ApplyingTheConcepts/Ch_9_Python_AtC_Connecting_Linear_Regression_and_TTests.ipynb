```python
#### Connecting Linear Regression and T-Tests ####

# Importing necessary libraries
from scipy import stats
import pandas as pd
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LinearRegression
from statsmodels.api import OLS
import statsmodels.api as sm

# Import data
carseats = pd.read_csv("carseats.csv")
carseats = carseats.dropna(subset = ["Sales", "US"])


sales_yes = carseats["Sales"][carseats["US"] == "Yes"]
sales_no = carseats["Sales"][carseats["US"] == "No"]

# T-Test
print("T-Test Output:")
stats.ttest_ind(sales_yes, sales_no, equal_var=True)
```

```python
# Construct regression model
carseats["USYes"] = 0
carseats["USYes"][carseats["US"] == "Yes"] = 1

lr_pipeline = Pipeline(
  [ ("lr", LinearRegression(fit_intercept = True))]
).set_output(transform="pandas")

X_train = pd.DataFrame(carseats["USYes"])
y_train = carseats["Sales"]
lr_fitted = lr_pipeline.fit(X_train, y_train)

## Traditional Linear Regression Output
print("Traditional Linear Regression T-statistic for US:")
X = sm.add_constant(X_train)
results = sm.OLS(y_train, X).fit()
print(results.tvalues[1])

print("Traditional Linear Regression P-value for US:")
print(results.pvalues[1])
```
