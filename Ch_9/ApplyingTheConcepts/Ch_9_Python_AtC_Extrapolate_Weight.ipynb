```python
#### Extrapolate Weight ####

# Importing necessary libraries
import pandas as pd
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LinearRegression

# Enter data from text into data frame
age = pd.DataFrame([1, 2, 3, 4, 6, 8, 10, 12, 14, 16, 18])
weight = [19, 29, 32, 37, 48, 62, 78, 98, 122, 145, 159]

# Build regression model
lr_pipeline = Pipeline(
  [("lr", LinearRegression(fit_intercept = True))]
).set_output(transform="pandas")

lr_fitted = lr_pipeline.fit(age, weight)

# Compute and print prediction
newage = pd.DataFrame([40])

print("Prediction at Age 40:")
print(lr_fitted.predict(newage))

# Extract the Linear Regression model from the pipeline
model = lr_fitted.named_steps['lr']

# Get the coefficient for 'age'
coefficient = model.coef_

# Get the intercept
intercept = model.intercept_

# Print coefficients
print("Regression Output:")
print("Coefficient (slope) for age:")
print(coefficient[0])
print("Intercept:")
print(intercept)
```
