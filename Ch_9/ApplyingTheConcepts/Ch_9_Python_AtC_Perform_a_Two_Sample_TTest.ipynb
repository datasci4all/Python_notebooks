```python
#### Perform a Two-Sample T-Test ####

# Importing necessary libraries
from scipy import stats
import pandas as pd

# Import data
carseats = pd.read_csv("carseats_sales_us.csv")

sales_yes = carseats["sales_us_yes"]
sales_no = carseats["sales_us_no"]

# T-Test
print("T-Test Output:")
stats.ttest_ind(sales_yes, sales_no, nan_policy='omit', equal_var=False)
```

```python
# Compute and print means
print("Sales for US Stores:")
print(sales_yes.mean())

print("Sales for non-US Stores:")
print(sales_no.mean())
```
